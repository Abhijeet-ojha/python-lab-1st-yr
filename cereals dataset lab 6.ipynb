```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
path="cereal.CSV"
df=pd.read_csv(path)
df.head()
```

```python
#a first part(boxplot)
plt.figure(figsize=(12,6))
sns.boxplot(data=df,orient='h',palette='plasma')
plt.title("boxplot for cereals datset")
plt.xlabel("values")
```

```python
#a secondpart(histplot)
plt.figure(figsize=(12,6))
sns.histplot(data=df,palette='viridis')
plt.xlabel('values')
plt.ylabel('Frequency')
plt.show()
```

```python
#b
highest_rating=df.loc[df['Rating'].idxmax()]
lowest_rating=df.loc[df['Rating'].idxmin()]
plt.figure(figsize=(12,6))
sns.barplot(x=['highest','lowest'],y=[highest_rating['Rating'],lowest_rating['Rating']],palette=['red','blue'])
plt.xlabel('cereal')
plt.ylabel('Rating')
```

```python
#c
plt.figure(figsize=(10, 6))
sns.regplot(x=df['Sugars'], y=df['Rating'], data=df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
```

```python
#d
avg_rating_by_manufacturer= df.groupby('Name')['Rating'].mean().reset_index().head(10)
avg_rating_by_manufacturer = avg_rating_by_manufacturer.sort_values(by='Rating', ascending=False)
plt.figure(figsize=(8, 6))
sns.barplot(x='Name', y='Rating', data=avg_rating_by_manufacturer, palette="Blues")
plt.xticks(rotation=45)
plt.show()
```
