```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# first import the ipl dataset
path=r'C:\Users\Abhijeet ojha\Downloads\ipl dataset.csv'
df=pd.read_csv(path)
```

```python
#q1
sns.boxplot(x='over',y='total_runs',data=df)
```

```python
#q2

top_10_batter = df.sort_values(by='total_runs', ascending=False).head(10)
plt.bar(top_10_batter['batter'],top_10_batter['total_runs'])
plt.xticks(rotation=45)
```

```python
#q3
frequency = df['extras_type'].value_counts()
sns.countplot(data=frequency)
```

```python
#q4
plt.figure(figsize=(14,8))
values=df['dismissal_kind'].value_counts()
plt.pie(values, startangle=90, autopct='%1.1f%%')  # Remove labels here
plt.legend(values.index, title="Dismissal Kind", loc="best")
plt.title('Dismissal Kind Frequency')
plt.show()

```

```python
#q5
economy=df['total_runs']/df['ball']
plt.scatter(economy, df['over'])
plt.xlabel('Economy')
plt.ylabel('Overs')
plt.title('Scatterplot from CSV')

plt.show()
```
