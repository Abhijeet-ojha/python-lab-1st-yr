```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
path="mtcars.csv"
df=pd.read_csv(path)
df.head()
```

```python
plt.figure(figsize=(12,8))
plt.hist(df['mpg'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel("miles per gallon")
plt.ylabel("Frequency")
plt.title("Distribution of Miles per Gallon")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df['hp'])

# Set title and labels
plt.title("Boxplot of horsepower in mtcars Dataset")
plt.xlabel("horsepower")
plt.ylabel("Values")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x="hp", y="mpg", data=df,hue="mpg",palette="viridis")
plt.title("Scatter Plot: hp vs mpgs")
plt.xlabel("hp")
plt.ylabel("mpg")

# Show plot
plt.show()
```

```python
average_mpg = df.groupby('cyl')['mpg'].mean().reset_index()
# Preview the average mpg
print(average_mpg)
# Create a bar plot for average mpg by cylinder category
plt.figure(figsize=(8, 6))
sns.barplot(x='cyl', y='mpg', data=average_mpg, palette="Blues")

# Set title and labels
plt.title('Average MPG by Cylinder Category')
plt.xlabel('Number of Cylinders')
plt.ylabel('Average MPG')

# Show the plot
plt.show()
```
