```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path=r'C:\Users\maheit\Downloads\sentiment_dataset.csv'
df=pd.read_csv(path)
```

```python
#q1
df["Text"]=df["Text"].fillna("no review provided")
```

```python
#q2
df.drop_duplicates()
```

```python
#q3
df["Sentiment"].map({"Positive":1,"Negative":0})
```

```python
#q4
unique=df["Text"].unique()
print(unique.size)
```

```python
#q5
df["Text"]=df["Text"].str.replace("bad","poor")
```

```python
#q6
longest_text = max(df['Text'], key=len)
shortest_text = min(df['Text'], key=len)
print("longest :",longest_text)
print("shortest :",shortest_text)
```

```python
#q7
plt.figure(figsize=(12,8))
freq=df['Sentiment'].value_counts()
freq.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel("Sentiment")
plt.ylabel("Frequency")
plt.show()
```
