```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# first import the ipl dataset
path=r'C:\Users\Abhijeet ojha\Downloads\ipl dataset.csv'
df=pd.read_csv(path)
```

```python
#q1
sns.boxplot(x='over',y='total_runs',data=df)
```

```python
#q2
top_10_batters = df.groupby('batter')['total_runs'].sum().nlargest(10).reset_index()
top_10_batters = top_10_batters.sort_values(by='total_runs', ascending=False)
plt.bar(top_10_batters['batter'],top_10_batters['total_runs'])
plt.xticks(rotation=45)
```

```python
#q3
sns.countplot(data=df,x=df['extras_type'])
```

```python
#q4
plt.figure(figsize=(14,8))
values=df['dismissal_kind'].value_counts()
plt.pie(values, startangle=90, autopct='%1.1f%%')  # Remove labels here
plt.legend(values.index, title="Dismissal Kind", loc="best")
plt.title('Dismissal Kind Frequency')
plt.show()

```

```python

#q5

# Calculate total runs and total balls per bowler
bowler_stats = df.groupby('bowler').agg({
    'total_runs': 'sum',
    'ball': 'count'  # Assuming one row per ball
}).reset_index()

# Calculate overs and economy
bowler_stats['overs'] = bowler_stats['ball'] / 6
bowler_stats['economy'] = bowler_stats['total_runs'] / bowler_stats['overs']

# Plot
plt.figure(figsize=(10,6))
plt.scatter(bowler_stats['economy'], bowler_stats['overs'], alpha=0.7)
plt.xlabel('Economy')
plt.ylabel('Overs Bowled')
plt.title('Bowler Economy vs Overs Bowled (IPL Dataset)')
plt.grid(True)
plt.show()
```
